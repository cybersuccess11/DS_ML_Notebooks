```python
import matplotlib.pyplot as plt

# Data for the bar plot
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [25, 18, 30, 12]

# Create the bar plot
plt.bar(categories, values, width=0.5, color='skyblue', align='center')

# Add labels to the bars
for i, value in enumerate(values):
    #plt.text(i, value + 1, str(value), ha='center', va='bottom')
    #plt.text(i, value + 1, str(value), va='bottom')
    plt.text(i, value + 1, str(value))

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.grid(True, axis='y')

# Show the plot
plt.show()
```

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np

# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30]) #Jan
y2 = np.array([20, 25, 15, 25]) #Feb
y3 = np.array([12, 15, 19, 6]) #March
y4 = np.array([10, 29, 13, 19]) #April

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np

# create data
x = ['Product A', 'Product B', 'Product C', 'Product D']
# Jan, Feb, March, April
y1 = np.array([10, 20, 10, 30]) 
y2 = np.array([20, 25, 15, 25]) 
y3 = np.array([12, 15, 19, 6]) 
y4 = np.array([10, 29, 13, 19]) 

# plot bars in stack manner
plt.barh(x, y1, color='r')
plt.barh(x, y2, left=y1, color='b')
plt.barh(x, y3, left=y1+y2, color='y')
plt.barh(x, y4, left=y1+y2+y3, color='g')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.legend(["Jan", "Feb", "Mar", "Apr"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = np.arange(5)
y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90]
y3 = [14, 23, 45, 25, 89]
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width, color='cyan')
plt.bar(x, y2, width, color='orange')
plt.bar(x+0.2, y3, width, color='green')
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])
plt.xlabel("Teams")
plt.ylabel("Scores")
plt.legend(["Round 1", "Round 2", "Round 3"])
plt.show()
```
