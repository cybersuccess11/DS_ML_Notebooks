```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
   'x': [1.2, 2.4, 3.5, 4.1, 5.0, 6.3, 7.2, 8.8, 9.5, 10.1],
   'y': [3.4, 5.1, 4.9, 6.2, 7.8, 9.0, 8.2, 10.5, 11.2, 12.3]
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Create the scatter plot
plt.scatter(df['x'], df['y'], c ="r", marker='^')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
line_x = np.linspace(0, 1, 100)
line_y = np.sin(line_x * np.pi)

# Create the plot
plt.plot(line_x, line_y, color='blue', label='Line Plot')
plt.scatter(x, y, color='red', marker='o', label='Scatter Plot')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot and Line Plot Example')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(50, 200, size=100)
colors = np.random.rand(100)

# Create the scatter plot
plt.scatter(x, y, s=sizes, c=colors, marker='s')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Different Marker Styles, Sizes, and Colors')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV data
df = pd.read_csv('car.csv')

# Extract the relevant columns
mileage = df['Mileage']
price = df['Price']
age = df['Age']

# Create the scatter plot
scatter = plt.scatter(mileage, price, c=age, cmap='coolwarm')

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Car Age')

# Add labels and title
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Car Mileage vs. Price with Car Age')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV data
df = pd.read_csv('car.csv')

# Extract the relevant columns
mileage = df['Mileage']
price = df['Price']
age = df['Age']

# Create the scatter plot
scatter = plt.scatter(mileage, price, c=age, cmap='rainbow')

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Car Age')

# Add labels and title
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Car Mileage vs. Price with Car Age')

# Show the plot
plt.show()
```
