### Plot for Linear, Quadratic and Cubic Functions

```python
import numpy as np
import matplotlib.pyplot as plt
# create a NumPy array of x values
x = np.arange(1,11)
# calculate the y values for the linear function
y_linear = x
# calculate the y values for the quadratic function
y_quadratic = x**2
# calculate the y values for the cubic function
y_cubic = x**3
print(y_linear)
print(y_quadratic)
print(y_cubic)
# plot the linear function
plt.plot(x, y_linear, label='Linear Func')
# # plot the quadratic function
plt.plot(x, y_quadratic, label='Quadratic Func')
# # plot the cubic function
plt.plot(x, y_cubic, label='Cubic Func')
# # add axis labels and a legend
plt.xlabel('vlaues')
plt.ylabel('resultant values post func application')
plt.legend()
# # show the plot
plt.show()
```

### Pandas' Plot method

```python
import pandas as pd
import matplotlib.pyplot as plt
# create a sample dataset
data = {'Year': [2010, 2011, 2012, 2013, 2014],
       'Sales': [100, 200, 300, 400, 500]}
# create a pandas DataFrame from the data
df = pd.DataFrame(data)
# plot the Sales data as a line plot
df.plot(x='Year', y='Sales', kind='line')
# display the plot
plt.show()
```

### Sample Plot using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# # Create a line plot
plt.plot(x, y)
# # Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot of Sine Function')
# # Display the plot
plt.show()
```

### plt.plot()

```python
import matplotlib.pyplot as plt

# create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# create a line plot
plt.plot(x, y)

# show the plot
plt.show()
```

### plt.bar()

```python
import matplotlib.pyplot as plt

# create some data
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 7, 15, 12, 8]

# create a bar plot
plt.bar(x, y)

# show the plot
plt.show()
```

### plt.hist()

```python
import matplotlib.pyplot as plt
import numpy as np

# create some data
data = np.random.normal(size=1000)

# create a histogram
plt.hist(data)

# show the plot
plt.show()
```

### plt.scatter()

```python
import matplotlib.pyplot as plt
import numpy as np

# create some data
x = np.random.normal(size=100)
y = np.random.normal(size=100)

# create a scatter plot
plt.scatter(x, y)

# show the plot
plt.show()
```

### plt.xlabel() and plt.ylabel()

```python
import matplotlib.pyplot as plt
# create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# create a line plot
plt.plot(x, y)
# set the x-axis label
plt.xlabel('X')
plt.ylabel('Y')
# show the plot
plt.show()
```
