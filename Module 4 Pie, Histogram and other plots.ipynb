### Pie Chart

```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
explode = [0, 0.1, 0, 0, 0]
colors = ['red', 'blue', 'green', 'yellow', 'orange']
autopct_format = '%.2f%%'

# Create a pie chart
plt.pie(sizes, labels=labels, explode=explode, colors=colors,
       autopct=autopct_format, shadow=True)

# Add a legend with a title
legend = plt.legend(title='Legend')

# Show the plot
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 3, 4, 4, 4, 5, 5, 6, 6, 7, 7, 7, 8, 9, 10]

# Create a histogram chart
plt.hist(data, bins=5, edgecolor='black')

# Customize the plot
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Chart')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)
data3 = np.random.normal(0, 3, 100)

# Create a figure and axes
fig, ax = plt.subplots()

# Create a box plot
ax.boxplot([data1, data2, data3])

# Set labels and title
ax.set_xticklabels(['Data 1', 'Data 2', 'Data 3'])
ax.set_xlabel('Datasets')
ax.set_ylabel('Values')
ax.set_title('Box Plot')

# Show the plot
plt.show()
```

### Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
data = np.random.rand(5, 5)  # Replace with your meaningful data

# Compute correlation matrix
corr_matrix = np.corrcoef(data)

# Plot the heatmap
plt.imshow(corr_matrix, cmap='coolwarm')
plt.colorbar()
plt.title('Correlation Heatmap')

# Set tick labels
attributes = ['Attribute 1', 'Attribute 2', 'Attribute 3', 'Attribute 4', 'Attribute 5']  # Replace with your attribute names
plt.xticks(range(len(attributes)), attributes, rotation=90)
plt.yticks(range(len(attributes)), attributes)

plt.show()
```
