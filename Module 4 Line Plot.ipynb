### Simple Line Plot along with labels

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the line chart
plt.plot(x, y)

# Set the title
plt.title("Line Chart")

# Set the x-axis label
plt.xlabel("X-axis")

# Set the y-axis label
plt.ylabel("Y-axis")

# Display the chart
plt.show()
```

### Markers

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the line chart
#plt.plot(x, y, marker='s')
#plt.plot(x, y, marker='D')
plt.plot(x, y, marker='^', mfc='r', mec='b', ms=15)

# Set the title
plt.title("Line Chart")

# Set the x-axis label
plt.xlabel("X-axis")

# Set the y-axis label
plt.ylabel("Y-axis")

# Display the chart
plt.show()
```

### Linestyles in Line Plot

```python
plt.plot(x, y, ls='dotted', lw=2, c = 'r')
plt.show()
```

### Font in Line plot

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Set the title with font properties
plt.title("Line Chart", fontfamily='Arial', fontsize=16, color='blue')

# Set the x-axis label with font properties
plt.xlabel("X-axis", fontfamily='Verdana', fontsize=10, color='red')

# Set the y-axis label with font properties
plt.ylabel("Y-axis", fontfamily='Verdana', fontsize=10, color='red')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

#font dict
font_dict = {'family':'Arial', 'size': 10, 'color': 'r'}

# Create a line plot
plt.plot(x, y)

# Set the title with font properties
plt.title("Line Chart", fontfamily='Arial', fontsize=16, color='blue')

# Set the x-axis label with font properties
plt.xlabel("X-axis", fontdict=font_dict)

# Set the y-axis label with font properties
plt.ylabel("Y-axis", fontdict=font_dict)

# Show the plot
plt.show()
```

### Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Sine Function')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Cosine Function')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Subplot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Subplot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y1)
plt.title('Subplot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y2)
plt.title('Subplot 4')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a line plot for y1
plt.plot(x, y1, label='Sine Function')

# Create a line plot for y2
plt.plot(x, y2, label='Cosine Function')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

y = [(y1, 'Sine'), (y2, 'Cosine')]
# Create a line plot for y1
for entry in y:
    plt.plot(x, entry[0], label= entry[1])

plt.legend()
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a line plot for y1
plt.plot(x, y1, label='Sine Function')

# Create a line plot for y2
plt.plot(x, y2, label='Cosine Function')

# Add a legend with customization
plt.legend(loc='upper right', fontsize='medium', frameon=True, shadow=True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y)

# Set x-axis limits
plt.xlim(0, 12)

# Set y-axis limits
plt.ylim(-1, 1)

# Set custom x-axis tick positions and labels
plt.xticks([0, 5, 10], ['Start', 'Middle', 'End'])

# Set custom y-axis tick positions and labels
plt.yticks([-1, 0, 1], ['Low', 'Mid', 'High'])

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Custom Axes and Ticks')

# Show the plot
plt.show()
```
